## Intro

### How does a KNN work?
##### میگه برای یک ایگرگ جدید که میخواهیم پیشبینی بکنیمش ببین نزدیک ترین کلاس براش کجا میتونه باشه (نزدیک ترین همسایه)
###### کمک میگریم و مثلا بهش میگیم پنج تا از نزدیک تریناشو پیدا کن و ببین ازین پنج تا تو کدوم کلاس بیشتر درگیرن K ولی یک مشکلی که این روش داره میتونه باعث زیاد شدن نویز ما بشه که اینجا ما از 
##### در این روش اهمیت زیادی دارع K پیدا کردن عدد مناسب برای 

### TIPS:
##### از این روش میشه برای هدف هایی که پیوسته هستند استفاده بکنیم 
##### فرض کنیم میخایم یک چیزی رو بسنجیم که مثلا عدد هست و مثلا پنج تا از نزدیک ترین حقوق که میانگین گرفتیم ازش

## KNN Evaluation 

### Different model evaluation metrics:

#### - Jaccard index
##### Tip: The easiest method in evaluating KNN 
##### Des: پیش بینی ما بر اساس مدل باشه (همون تست و تمرین) این روش میگه بیا شباهت این دوتارو حساب کن  ŷ پیشبینی درست باشه و  y درواقع این متود میگه اگر 
##### Guide: 

#### - F1-score 
##### Tip: درواقع متودی هست برای کامل تر کردن متود جاکارد ایندکس
##### Tip: این متود میگه برای من مهمه که تو چی گفتی و چی شد ولی جاکارد میگه تو توی چند درصد موارد درست گقتی 
##### *TP* => من گفتم یارو میره و رفته (انتظار ضرر داشتم و ضرر هم خوردم)
##### *FN* => من گفتم یارو نمیره ولی رفته (انتظار ضرر نداشتم ولی ضرر خوردم)
##### *FP* => من گفتم یارو میره ولی نرفته (انتظار ضرر داشتم ولی سود آوردم)
##### *TN* => من گفتم یارو نمیره و نرفته (انتظار ضرر نداشتم و ضرر هم ندیدم)
##### Tip => سودآور ترین هست FP هستش و خیلی هاد هست و درمقابل FN بیشترین ضرر رو به ما درواقع
##### Precision => شما چقدر دقیق پیشبینی کردید
##### Recall => چندتا ازونایی که خوندید رو درست پیشبینی کردید
