## Hands On

```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn import preprocessing, metrics
from sklearn.model_selection import train_test_split
%matplotlib inline
```

```python
df = pd.read_csv("../00-storage/csv-store/drug200.csv", delimiter=",")
df[0:5]
```

## Pre Processing 

```python
x = df[["Age", "Sex", "BP", "Cholesterol", "Na_to_K"]].values
x[0:5]
```

##### convert Sex and BP and Cholesteril from categorical to numerical 

```python
le_sex = preprocessing.LabelEncoder()
le_sex.fit(["F", "M"])
x[:,1] = le_sex.transform(x[:,1])

le_BP = preprocessing.LabelEncoder()
le_BP.fit(["LOW", "NORMAL", "HIGH"])
x[:,2] = le_BP.transform(x[:,2])

le_Chol = preprocessing.LabelEncoder()
le_Chol.fit(["NORMAL", "HIGH"])
x[:,3] = le_Chol.transform(x[:,3])

x[0:5]
```

```python
y = df["Drug"]
y[0:5]
```

## Setting Up The Decision Tree 

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=3)
```

## Modeling

```python
drugTree = DecisionTreeClassifier(criterion="entropy", max_depth = 4)
drugTree
```

```python
drugTree.fit(x_train, y_train)
```

## Prediction

```python
yhat = drugTree.predict(x_test)
```

```python
print(yhat[0:5])
print(y_test[0:5])
```

## Evaluation

```python
print("DecitionTree's Accuracy: ", metrics.accuracy_score(y_test, yhat))
```

## Vizualization

```python
from io import StringIO
import pydotplus
import matplotlib.image as mpimg
from sklearn import tree
```

```python
dot_data = StringIO()
filename = "drugtree.png"
featureNames = df.columns[0:5]

out = tree.export_graphviz(
    drugTree,
    feature_names=featureNames,
    out_file=dot_data,
    class_names= np.unique(y_train),
    filled=True,
    special_characters=True,
    rotate=False
)

graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
graph.write_png(filename)

img = mpimg.imread(filename)

plt.figure(figsize=(100,200))
plt.imshow(img, interpolation="nearest")
plt.show()
```
